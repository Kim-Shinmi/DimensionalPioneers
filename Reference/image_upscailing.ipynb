# Result 이미지 upscailing

```python
import cv2
import numpy as np

# 이미지 파일 경로 리스트 생성
image_paths = [f'/content/drive/MyDrive/TeamProjects/DimensionalPioneers/4.zero123/Result/output{i}.png' for i in range(1, 7)]

# 각 이미지에 대해서 업스케일링을 수행
for i, image_path in enumerate(image_paths):
    # 이미지를 읽어옴
    image = cv2.imread(image_path)

    # 이미지 업스케일링 비율 설정
    scale_percent = 400  # 400% 크기로 설정

    # 새로운 이미지 크기 계산
    width = int(image.shape[1] * scale_percent / 100)
    height = int(image.shape[0] * scale_percent / 100)
    dim = (width, height)

    # 이미지 업스케일링
    resized = cv2.resize(image, dim, interpolation=cv2.INTER_CUBIC)

    # 업스케일링된 이미지를 새 파일 이름으로 저장
    upscaled_image_path = f'/content/drive/MyDrive/TeamProjects/DimensionalPioneers/4.zero123/Result/output{i+1}.png'
    cv2.imwrite(upscaled_image_path, resized)

    # 결과 확인
    print(f'Image {i+1} upscaled to:', resized.shape)
```

# 조명 조정

```python
import cv2
import numpy as np

# 이미지 로드
image_paths = [f'/content/drive/MyDrive/TeamProjects/DimensionalPioneers/4.zero123/Result/output{i}.png' for i in range(1, 7)]

# 각 이미지에 대해서 조명 조정을 수행
for i, image_path in enumerate(image_paths):
    # 이미지를 읽어옴
    image = cv2.imread(image_path)

    # 대비와 밝기 조정 파라미터
    alpha = 1.2 # 대비 조정 (1.0은 원본, 1.0보다 크면 대비 증가)
    beta = 50   # 밝기 조정 (양수는 밝게, 음수는 어둡게)

    # 대비와 밝기 조정 적용
    adjusted_image = cv2.convertScaleAbs(image, alpha=alpha, beta=beta)

    # 결과 이미지 저장 경로
    adjusted_image_path = f'/content/drive/MyDrive/TeamProjects/DimensionalPioneers/4.zero123/Result/output{i+1}.png'
    cv2.imwrite(adjusted_image_path, adjusted_image)

    # 결과 확인
    print(f'Image {i+1} adjusted and saved to:', adjusted_image_path)
```

# 잡음 제거

```python
import cv2
import numpy as np

# 이미지 로드
image_paths = [f'/content/drive/MyDrive/TeamProjects/DimensionalPioneers/4.zero123/Result/output{i}.png' for i in range(1, 7)]

# 각 이미지에 대해서 잡음 제거를 수행
for i, image_path in enumerate(image_paths):
    # 이미지를 읽어옴
    image = cv2.imread(image_path)

    # Gaussian Blur 적용
    # (5, 5)는 Gaussian Kernel의 크기, 0은 표준 편차를 자동으로 계산하게 함
    denoised_image = cv2.GaussianBlur(image, (5, 5), 0)

    # 결과 이미지 저장 경로
    denoised_image_path = f'/content/drive/MyDrive/TeamProjects/DimensionalPioneers/4.zero123/Result/output{i+1}.png'
    cv2.imwrite(denoised_image_path, denoised_image)

    # 결과 확인
    print(f'Image {i+1} denoised and saved to:', denoised_image_path)
```

# 이미지 샤프닝

```python
import cv2
import numpy as np

# 이미지 로드
image_paths = [f'/content/drive/MyDrive/TeamProjects/DimensionalPioneers/4.zero123/Result/output{i}.png' for i in range(1, 7)]

# 각 이미지에 대해서 샤프닝을 수행
for i, image_path in enumerate(image_paths):
    # 이미지를 읽어옴
    image = cv2.imread(image_path)

    # 샤프닝을 위한 커널 정의
    kernel_sharpening = np.array([[-1, -1, -1],
                                  [-1,  9, -1],
                                  [-1, -1, -1]])

    # 이미지에 샤프닝 적용
    sharpened_image = cv2.filter2D(image, -1, kernel_sharpening)

    # 결과 이미지 저장 경로
    sharpened_image_path = f'/content/drive/MyDrive/TeamProjects/DimensionalPioneers/4.zero123/Result/output{i+1}.png'
    cv2.imwrite(sharpened_image_path, sharpened_image)

    # 결과 확인
    print(f'Image {i+1} sharpened and saved to:', sharpened_image_path)
```

# 이미지 크롭핑 및 회전
 - 크롭핑: 이미지에서 관심 영역(Region of Interest, ROI)을 강조하여 특징점 추출의 정확도를 높이고, 불필요한 정보로 인한 혼란을 줄일 수 있습니다.
 - 회전: 서로 다른 방향으로 촬영된 이미지들 사이의 특징점 매칭을 개선하기 위해 이미지의 방향을 조정합니다. 이는 특히, 같은 객체나 장면을 다양한 각도에서 촬영한 경우 유용합니다.

```python
import cv2
import numpy as np

# 이미지 로드
image_paths = [f'/content/drive/MyDrive/TeamProjects/DimensionalPioneers/4.zero123/Result/output{i}.png' for i in range(1, 7)]

# 각 이미지에 대해서 크롭 및 회전을 수행
for i, image_path in enumerate(image_paths):
    # 이미지를 읽어옴
    image = cv2.imread(image_path)

    # 이미지 크롭 (관심 영역 정의)
    # 여기서는 이미지의 중앙 부분을 크롭한다고 가정
    h, w = image.shape[:2]
    cropped_image = image[int(h/4):int(3*h/4), int(w/4):int(3*w/4)]

    # 이미지 90도 회전
    rotated_image = cv2.rotate(cropped_image, cv2.ROTATE_90_CLOCKWISE)

    # 결과 이미지 저장 경로
    processed_image_path = f'/content/drive/MyDrive/TeamProjects/DimensionalPioneers/4.zero123/Result/processed_output{i+1}.png'
    cv2.imwrite(processed_image_path, rotated_image)

    # 결과 확인
    print(f'Image {i+1} cropped and rotated, saved to:', processed_image_path)
```

# 동일한 크기 및 비율 조정

# 색상 공간 변환

```python
import cv2
import numpy as np

# 이미지 로드
image_paths = [f'/content/drive/MyDrive/TeamProjects/DimensionalPioneers/4.zero123/Result/output{i}.png' for i in range(1, 7)]

# 각 이미지에 대해서 색상 공간을 그레이스케일로 변환
for i, image_path in enumerate(image_paths):
    # 이미지를 읽어옴
    image = cv2.imread(image_path)

    # 그레이스케일로 색상 공간 변환
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # 결과 이미지 저장 경로
    gray_image_path = f'/content/drive/MyDrive/TeamProjects/DimensionalPioneers/4.zero123/Result/output{i+1}.png'
    cv2.imwrite(gray_image_path, gray_image)

    # 결과 확인
    print(f'Image {i+1} converted to grayscale, saved to:', gray_image_path)
```
